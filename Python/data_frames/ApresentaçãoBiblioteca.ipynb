```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('https://raw.githubusercontent.com/Ingrid27/dataframeNotas/ac6bd2a879ed4c24d40f3fe932193dfd86bc55b1/frameNotas.csv', sep = ';')
```

```python
df
```

```python
graphic1 = df['Sexo '].value_counts()

graphic1 = graphic1.reset_index()
graphic1 = graphic1.rename(columns={'Sexo ':'Quantidade'})
graphic1 = graphic1.rename(columns={'index':'Sexo'})

fig = px.pie(graphic1, names= 'Sexo', values='Quantidade')

fig.show()
graphic1
```

```python
graphic2 = df['Notas'].value_counts()

graphic2 = graphic2.reset_index()
graphic2 = graphic2.rename(columns={'Notas':'Quantidade de Notas'})
graphic2 = graphic2.rename(columns={'index':'Notas'})

fig2 = px.bar(graphic2, x= 'Notas', y='Quantidade de Notas')

fig2.show()
graphic2
```

```python
notaTurma= df[['Nome', 'Turma', 'Notas']]

notaTurma = notaTurma.loc[notaTurma['Notas']>=8,['Turma', 'Notas']]

notaTurma = notaTurma.sort_values('Notas')

notaTurma = notaTurma.value_counts().reset_index()

notaTurma = notaTurma.rename(columns={0:'Quantidade de Notas'})

figTurma = px.bar(notaTurma, y= 'Quantidade de Notas', x= 'Turma', color='Notas')

figTurma.show()
notaTurma
```
