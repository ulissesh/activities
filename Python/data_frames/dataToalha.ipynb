```python
graphic1 = df['produto'].value_counts()

graphic1 = graphic1.reset_index()

fig = px.bar(graphic1, x= 'index', y='produto')

fig.show()
```

```python
def transforma_minuscula(x):
  return(str(x).lower())
df['setor']=df['setor'].apply(func = transforma_minuscula)

```

```python
df
```

```python
df['TotaL'] = df['TotaL'].map(lambda x: str(x).replace('R$ ', ''))
```

```python
df['valor UN'] = df['valor UN'].map(lambda x: str(x).replace(',', '.'))
```

```python
df['valor UN'] = pd.to_numeric(df['valor UN'])

```
